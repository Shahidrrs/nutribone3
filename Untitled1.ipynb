```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
import numpy as np
import os
import shutil
import random
import seaborn as sns
```

```python
import kagglehub

print("Downloading dataset 1 (mohamedgobara)...")
path1 = kagglehub.dataset_download("mohamedgobara/multi-class-knee-osteoporosis-x-ray-dataset")
print("Dataset 1 path:", path1)

print("\nDownloading dataset 2 (stevepython)...")
path2 = kagglehub.dataset_download("stevepython/osteoporosis-knee-xray-dataset")
print("Dataset 2 path:", path2)
```

```python
import random
import shutil
import os
import numpy as np
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

random.seed(42)
np.random.seed(42)

# ── Delete old merged folder ──
if os.path.exists("/tmp/osteoporosis_merged"):
    shutil.rmtree("/tmp/osteoporosis_merged")
    print("Old merged folder deleted.")

merged_dir = "/tmp/osteoporosis_merged"
splits     = ['train', 'val', 'test']
classes    = ['Normal', 'Osteopenia', 'Osteoporosis']

# Exact targets — all classes perfectly equal
targets = {
    'train': 1500,
    'val':   300,
    'test':  150
}

# Create folder structure
for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(merged_dir, split, cls), exist_ok=True)

def collect_images(src_dir):
    if not os.path.exists(src_dir):
        print(f"  WARNING: Not found: {src_dir}")
        return []
    return [
        os.path.join(src_dir, f)
        for f in os.listdir(src_dir)
        if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))
    ]

def copy_images(image_paths, dst_dir, label_prefix):
    """Copy images into dst_dir with unique names."""
    count = 0
    for i, src in enumerate(image_paths):
        ext = os.path.splitext(src)[1].lower()
        if ext not in ['.jpg', '.jpeg', '.png', '.bmp']:
            ext = '.jpg'
        dst = os.path.join(dst_dir, f"{label_prefix}_{i}{ext}")
        shutil.copy2(src, dst)
        count += 1
    return count

def augment_to_exact(src_dir, target_count, label_prefix='aug'):
    """
    Augment or trim images in src_dir to reach exactly target_count.
    If more images than target: randomly sample down.
    If fewer images than target: augment up.
    """
    existing = [
        os.path.join(src_dir, f)
        for f in os.listdir(src_dir)
        if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))
    ]
    current = len(existing)

    # If we have MORE than target — randomly remove excess
    if current > target_count:
        to_remove = random.sample(existing, current - target_count)
        for f in to_remove:
            os.remove(f)
        print(f"  Trimmed {current} → {target_count} (removed {current - target_count})")
        return

    # If we have FEWER than target — augment up
    if current < target_count:
        aug = ImageDataGenerator(
            rotation_range=20,
            width_shift_range=0.1,
            height_shift_range=0.1,
            zoom_range=0.15,
            horizontal_flip=True,
            brightness_range=[0.8, 1.2],
            shear_range=0.1,
            fill_mode='nearest'
        )

        needed = target_count - current
        idx    = 0
        print(f"  Augmenting {current} → {target_count} (need {needed} more)...")

        while idx < needed:
            src_img_path = random.choice(existing)
            try:
                img = load_img(src_img_path, target_size=(224, 224))
                x   = img_to_array(img)
                x   = x.reshape((1,) + x.shape)
                for batch in aug.flow(x, batch_size=1):
                    aug_img = batch[0].astype(np.uint8)
                    save_path = os.path.join(src_dir, f"{label_prefix}_{idx}.jpg")
                    Image.fromarray(aug_img).save(save_path)
                    idx += 1
                    break
            except Exception as e:
                print(f"  Skipping image due to error: {e}")
                continue

        print(f"  Done. Total: {len(os.listdir(src_dir))} images")

    if current == target_count:
        print(f"  Already at target: {target_count} images")

# ── Dataset source paths ──
d1_base    = os.path.join(path1, "OS Collected Data")
d1_sources = {
    'Normal':       os.path.join(d1_base, "Normal"),
    'Osteopenia':   os.path.join(d1_base, "Osteopenia"),
    'Osteoporosis': os.path.join(d1_base, "Osteoporosis"),
}
d2_sources = {
    'Normal':       os.path.join(path2, "normal", "normal"),
    'Osteoporosis': os.path.join(path2, "osteoporosis", "osteoporosis"),
}

print("=" * 55)
print("Step 1: Collecting and splitting all real images...")
print("=" * 55)

# Store split counts for reporting
all_real = {}

for cls in classes:
    imgs_d1 = collect_images(d1_sources.get(cls, ""))
    imgs_d2 = collect_images(d2_sources.get(cls, ""))
    all_imgs = imgs_d1 + imgs_d2
    random.shuffle(all_imgs)

    n       = len(all_imgs)
    n_train = int(n * 0.70)
    n_val   = int(n * 0.20)
    n_test  = n - n_train - n_val

    split_map = {
        'train': all_imgs[:n_train],
        'val':   all_imgs[n_train:n_train + n_val],
        'test':  all_imgs[n_train + n_val:]
    }

    all_real[cls] = {}
    for split_name, paths in split_map.items():
        dst = os.path.join(merged_dir, split_name, cls)
        n_copied = copy_images(paths, dst, label_prefix='real')
        all_real[cls][split_name] = n_copied

    print(f"\n{cls}: {len(imgs_d1)} from D1 + {len(imgs_d2)} from D2 = {n} real total")
    for s in splits:
        print(f"  {s}: {all_real[cls][s]} real images")

print("\n" + "=" * 55)
print("Step 2: Augmenting/trimming all classes to exact targets...")
print("=" * 55)

for cls in classes:
    print(f"\n{cls}:")
    for split_name in splits:
        target    = targets[split_name]
        split_dir = os.path.join(merged_dir, split_name, cls)
        current   = len(os.listdir(split_dir))
        print(f"  {split_name}: {current} real → target {target}")
        augment_to_exact(split_dir, target, label_prefix=f'aug_{cls[:3].lower()}')

print("\n" + "=" * 55)
print("Final balanced dataset summary:")
print("=" * 55)
grand_total = 0
for split in splits:
    print(f"\n{split}/")
    for cls in classes:
        n = len(os.listdir(os.path.join(merged_dir, split, cls)))
        print(f"  {cls}: {n} images")
        grand_total += n

print(f"\nGrand total: {grand_total} images")
print(f"Expected:    {sum(targets.values()) * len(classes)} images")

# Verify all equal
print("\nVerification — all classes equal per split:")
all_equal = True
for split in splits:
    counts = [len(os.listdir(os.path.join(merged_dir, split, cls))) for cls in classes]
    equal  = len(set(counts)) == 1
    status = "✓" if equal else "✗"
    print(f"  {split}: {counts} {status}")
    if not equal:
        all_equal = False

if all_equal:
    print("\nAll classes perfectly balanced!")
else:
    print("\nWarning: some classes are not equal — recheck targets.")
```

```python
train_dir = os.path.join(merged_dir, 'train')
val_dir   = os.path.join(merged_dir, 'val')
test_dir  = os.path.join(merged_dir, 'test')

train_datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.85, 1.15],
    fill_mode='nearest',
    preprocessing_function=tf.keras.applications.vgg19.preprocess_input
)

val_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.vgg19.preprocess_input
)

test_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.vgg19.preprocess_input
)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=16,
    classes=['Normal', 'Osteopenia', 'Osteoporosis'],
    class_mode='categorical',
    shuffle=True
)

val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=16,
    classes=['Normal', 'Osteopenia', 'Osteoporosis'],
    class_mode='categorical',
    shuffle=False
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=1,
    classes=['Normal', 'Osteopenia', 'Osteoporosis'],
    class_mode='categorical',
    shuffle=False
)

print(f"Train : {train_gen.samples} images")
print(f"Val   : {val_gen.samples} images")
print(f"Test  : {test_gen.samples} images")
print(f"Classes: {train_gen.class_indices}")
```

```python
print("\nBuilding VGG-19 model...")

inputs = layers.Input(shape=(224, 224, 3))

base_model = tf.keras.applications.VGG19(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
)

# Stage 1: freeze entire backbone
base_model.trainable = False

x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.5)(x)

x = layers.Dense(512, activation='relu',
                  kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)

x = layers.Dense(256, activation='relu',
                  kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)

outputs = layers.Dense(3, activation='softmax')(x)

model = models.Model(inputs, outputs)
model.summary()
print("\nStage 1 ready — VGG-19 backbone frozen.")
```

```python
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=np.array([0, 1, 2]),
    y=train_gen.classes
)
class_weights = dict(enumerate(class_weights_array))
print("Class weights:", class_weights)

callbacks_s1 = [
    EarlyStopping(monitor='val_loss', patience=8,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                      patience=3, min_lr=1e-7, verbose=1),
    ModelCheckpoint("best_stage1.keras", monitor='val_accuracy',
                    save_best_only=True, verbose=1)
]

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nStage 1: Training head with frozen backbone...")
history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=30,
    callbacks=callbacks_s1,
    class_weight=class_weights
)

print(f"\nBest Stage 1 Val Accuracy: {max(history1.history['val_accuracy']):.4f}")
```

```python
print("\nStage 2: Unfreezing block4 + block5 of VGG-19...")

base_model.trainable = True
for layer in base_model.layers:
    layer.trainable = ('block5' in layer.name or
                       'block4' in layer.name)

print("Trainable layers:", [l.name for l in base_model.layers if l.trainable])

callbacks_s2 = [
    EarlyStopping(monitor='val_loss', patience=8,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                      patience=4, min_lr=1e-9, verbose=1),
    ModelCheckpoint("best_stage2.keras", monitor='val_accuracy',
                    save_best_only=True, verbose=1)
]

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=5e-6),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nStage 2: Fine-tuning block4 + block5...")
history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=40,
    callbacks=callbacks_s2,
    class_weight=class_weights
)

model.save("nutribone_vgg19_3class_final.keras")
print(f"\nBest Stage 2 Val Accuracy: {max(history2.history['val_accuracy']):.4f}")
```

```python
total_train_loss = history1.history['loss'] + history2.history['loss']
total_val_loss   = history1.history['val_loss'] + history2.history['val_loss']
total_train_acc  = history1.history['accuracy'] + history2.history['accuracy']
total_val_acc    = history1.history['val_accuracy'] + history2.history['val_accuracy']
split_point      = len(history1.history['loss']) - 1

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(total_train_loss, label='Train Loss', color='royalblue')
plt.plot(total_val_loss, label='Val Loss', color='orange', linestyle='--')
plt.axvline(x=split_point, color='gray', linestyle=':', label='Fine-Tune Split')
plt.legend()
plt.title('Loss (Stage 1 + Stage 2)')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
plt.plot(total_train_acc, label='Train Accuracy', color='royalblue')
plt.plot(total_val_acc, label='Val Accuracy', color='orange', linestyle='--')
plt.axvline(x=split_point, color='gray', linestyle=':', label='Fine-Tune Split')
plt.legend()
plt.title('Accuracy (Stage 1 + Stage 2)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
print("\nLoading best model...")
eval_model = tf.keras.models.load_model("best_stage2.keras")  # fixed

test_gen.reset()
preds  = eval_model.predict(test_gen, steps=len(test_gen), verbose=1)
y_pred = np.argmax(preds, axis=1)
y_true = test_gen.classes
class_names = ['Normal', 'Osteopenia', 'Osteoporosis']

print("\n--- Classification Report ---")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names,
            cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix — Test Set')
plt.show()

print(f"\nFinal Test Accuracy: {np.sum(y_pred == y_true) / len(y_true) * 100:.2f}%")
```

```python
print("\nApplying TTA (20 rounds)...")

tta_datagen = ImageDataGenerator(
    rotation_range=5,
    horizontal_flip=True,
    zoom_range=0.03,
    preprocessing_function=tf.keras.applications.vgg19.preprocess_input
)

tta_rounds = 20
tta_preds  = np.zeros((test_gen.samples, 3))

for i in range(tta_rounds):
    tta_gen = tta_datagen.flow_from_directory(
        test_dir,
        target_size=(224, 224),
        batch_size=1,
        classes=['Normal', 'Osteopenia', 'Osteoporosis'],
        class_mode='categorical',
        shuffle=False
    )
    tta_gen.reset()
    tta_preds += eval_model.predict(tta_gen, steps=len(tta_gen), verbose=0)
    print(f"TTA round {i+1}/{tta_rounds} done")

tta_preds /= tta_rounds
y_pred_tta = np.argmax(tta_preds, axis=1)

print("\n--- TTA Classification Report ---")
print(classification_report(y_true, y_pred_tta, target_names=class_names))

cm_tta = confusion_matrix(y_true, y_pred_tta)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_tta, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names,
            cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix — With TTA (20 rounds)')
plt.show()

print(f"\nFinal TTA Accuracy: {np.sum(y_pred_tta == y_true) / len(y_true) * 100:.2f}%")
```
