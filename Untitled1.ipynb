```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
import kagglehub

path = kagglehub.dataset_download("fuyadhasanbhoyan/knee-osteoarthritis-classification-224224")
print("Path to dataset files:", path)
```

```python
base_dataset_folder = os.path.join(path, "Knee Osteoarthritis Classification")

train_dir = os.path.join(base_dataset_folder, "train")
val_dir = os.path.join(base_dataset_folder, "val")
test_dir = os.path.join(base_dataset_folder, "test")

print(f"Training path: {train_dir}")
print(f"Validation path: {val_dir}")
print(f"Test path: {test_dir}")
```

```python
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    brightness_range=[0.8, 1.2],
    zoom_range=0.15,
    horizontal_flip=True,
    vertical_flip=False,
    shear_range=0.1,
    fill_mode='nearest',
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

val_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

test_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    classes=['Normal', 'Osteoporosis'],
    class_mode='binary',
    shuffle=True
)

val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    classes=['Normal', 'Osteoporosis'],
    class_mode='binary',
    shuffle=False
)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=1,
    classes=['Normal', 'Osteoporosis'],
    class_mode='binary',
    shuffle=False
)

print(f"\nTraining samples: {train_gen.samples}")
print(f"Validation samples: {val_gen.samples}")
print(f"Test samples: {test_gen.samples}")
print(f"Class indices: {train_gen.class_indices}")
```

```python
print("\nBuilding EfficientNetB3 model...")

inputs = layers.Input(shape=(224, 224, 3))

base_model = tf.keras.applications.EfficientNetB3(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
)

# Stage 1: Freeze entire backbone
base_model.trainable = False

x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.5)(x)

x = layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)

x = layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)

outputs = layers.Dense(1, activation='sigmoid')(x)

model = models.Model(inputs, outputs)
model.summary()
print("\nStage 1 model ready — backbone fully frozen.")
```

```python
# Class imbalance weights
counter = train_gen.classes
total = len(counter)
n_normal = np.sum(counter == 0)
n_osteo = np.sum(counter == 1)
class_weights = {
    0: total / (2 * n_normal),
    1: total / (2 * n_osteo)
}
print("Class weights:", class_weights)

callbacks_stage1 = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-7, verbose=1),
    ModelCheckpoint("best_stage1.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
]

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("\nStage 1: Training head with frozen backbone...")
history1 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=15,
    callbacks=callbacks_stage1,
    class_weight=class_weights
)

print(f"\nBest Stage 1 Val Accuracy: {max(history1.history['val_accuracy']):.4f}")
```

```python
print("\nStage 2: Unfreezing top 40% of backbone layers...")

base_model.trainable = True

# Freeze bottom 60%, unfreeze top 40%
total_layers = len(base_model.layers)
freeze_until = int(total_layers * 0.6)
for layer in base_model.layers[:freeze_until]:
    layer.trainable = False

trainable_count = sum(1 for l in base_model.layers if l.trainable)
print(f"Total backbone layers: {total_layers}")
print(f"Frozen layers: {freeze_until}")
print(f"Trainable backbone layers: {trainable_count}")

callbacks_stage2 = [
    EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-8, verbose=1),
    ModelCheckpoint("best_stage2.keras", monitor='val_accuracy', save_best_only=True, verbose=1)
]

model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-5, weight_decay=1e-5),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("\nStage 2: Fine-tuning with partial backbone unfreeze...")
history2 = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=30,
    callbacks=callbacks_stage2,
    class_weight=class_weights
)

model.save("nutribone_binary_effnetB3_final.keras")
print(f"\nBest Stage 2 Val Accuracy: {max(history2.history['val_accuracy']):.4f}")
print("Final model saved!")
```

```python
total_train_loss = history1.history['loss'] + history2.history['loss']
total_val_loss   = history1.history['val_loss'] + history2.history['val_loss']
total_train_acc  = history1.history['accuracy'] + history2.history['accuracy']
total_val_acc    = history1.history['val_accuracy'] + history2.history['val_accuracy']
split_point      = len(history1.history['loss']) - 1

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(total_train_loss, label='Train Loss', color='royalblue')
plt.plot(total_val_loss, label='Val Loss', color='orange', linestyle='--')
plt.axvline(x=split_point, color='gray', linestyle=':', label='Fine-Tune Split')
plt.legend()
plt.title('Loss Progression (Stage 1 + Stage 2)')
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
plt.plot(total_train_acc, label='Train Accuracy', color='royalblue')
plt.plot(total_val_acc, label='Val Accuracy', color='orange', linestyle='--')
plt.axvline(x=split_point, color='gray', linestyle=':', label='Fine-Tune Split')
plt.legend()
plt.title('Accuracy Progression (Stage 1 + Stage 2)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
print("\nLoading best saved model for evaluation...")
eval_model = tf.keras.models.load_model("best_stage2.keras")

# Reset test generator
test_gen.reset()

print("Running predictions on test set...")
preds = eval_model.predict(test_gen, steps=len(test_gen), verbose=1)

y_pred = (preds > 0.5).astype(int).flatten()
y_true = test_gen.classes
class_names = ['Normal', 'Osteoporosis']

print("\n--- Classification Report (Test Set) ---")
print(classification_report(y_true, y_pred, target_names=class_names))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names,
            cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix — Test Set')
plt.show()

# Overall accuracy
test_acc = np.sum(y_pred == y_true) / len(y_true)
print(f"\nFinal Test Accuracy: {test_acc * 100:.2f}%")
```

```python
print("\nApplying Test Time Augmentation (TTA) for final accuracy boost...")

tta_datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    preprocessing_function=tf.keras.applications.efficientnet.preprocess_input
)

tta_rounds = 10
tta_preds = np.zeros((test_gen.samples,))

for i in range(tta_rounds):
    tta_gen = tta_datagen.flow_from_directory(
        test_dir,
        target_size=(224, 224),
        batch_size=1,
        classes=['Normal', 'Osteoporosis'],
        class_mode='binary',
        shuffle=False
    )
    tta_gen.reset()
    p = eval_model.predict(tta_gen, steps=len(tta_gen), verbose=0)
    tta_preds += p.flatten()
    print(f"TTA round {i+1}/{tta_rounds} complete")

tta_preds /= tta_rounds
y_pred_tta = (tta_preds > 0.5).astype(int)

print("\n--- TTA Classification Report ---")
print(classification_report(y_true, y_pred_tta, target_names=class_names))

cm_tta = confusion_matrix(y_true, y_pred_tta)
plt.figure(figsize=(7, 5))
sns.heatmap(cm_tta, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names,
            cmap='Greens')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix — With TTA')
plt.show()

tta_acc = np.sum(y_pred_tta == y_true) / len(y_true)
print(f"\nFinal TTA Test Accuracy: {tta_acc * 100:.2f}%")
```
